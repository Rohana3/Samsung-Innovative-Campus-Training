```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(datalink)

df = df.reset_index(drop=True)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
cluster_df = df.copy()

cluster_df.head()
```

# KMeans

```python
cluster = KMeans(n_clusters=3, random_state=0)

cluster.fit(cluster_df)
```

```python
cluster.fit_predict(cluster_df)
```

```python
df['cluster'] = cluster.fit_predict(cluster_df)

df.head()
```

```python
df['cluster'].value_counts()
```

```python
df['cluster'].value_counts().hist()
```

```python
import seaborn as sns

sns.pairplot(df, hue='cluster',palette='tab10')
```

```python
sns.boxplot(x=df['Malic acid'],y=df['Alcohol'])
```

```python
sns.heatmap(cluster_df.corr(), annot=True)
```

```python
# help(sns)
```

```python
sns.scatterplot(x=df['Alcohol'],
            y=df['Malic acid'],
            hue=df['cluster'],
            palette='tab10')
```

# Heirarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hcluster = AgglomerativeClustering(n_clusters=3)
```

```python
hcluster.fit_predict(cluster_df)
```

```python
df['hcluster'] = hcluster.fit_predict(cluster_df)

df.sample(2)
```

```python
df['hcluster'].value_counts()
```

```python
sns.scatterplot(x=df['Alcohol'],
            y=df['Malic acid'],
            hue=df['hcluster'],
            palette='tab10')
```

# DBScan

```python
from sklearn.cluster import DBSCAN
```

```python
dcluster = DBSCAN(eps=0.5, min_samples=5)
```

```python
dcluster.fit_predict(cluster_df)
```

```python
df['dcluster'] = dcluster.fit_predict(cluster_df)

df.sample(2)
```

```python
df['dcluster'].value_counts()
```

```python
sns.scatterplot(x=df['Alcohol'],
            y=df['Malic acid'],
            hue=df['dcluster'],
            palette='tab10')
```
